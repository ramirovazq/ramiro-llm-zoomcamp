```python
from minsearch import Index
import minsearch
```

```python
import json
```

```python
with open('parse/documents.json', 'rt') as f_in:
    docs_raw = json.load(f_in)
    
    
```

```python
# {
#     'text': 'When you set up your accounr certificate!!!', 
#     'section': 'Course Management Platform for Homeworks, Project and Certificate', 
#     'question': 'Leaderboard - I am not on the leaderboard / how do I know which one I am on the leaderboard?', 
#     'course': 'data-engineering-zoomcamp'
# }

documents = []
for dict_course in docs_raw:
    course = dict_course['course']
    for dict_document in dict_course['documents']:
        documents.append({ 
            'course': course,
            'text': dict_document['text'],
            'section': dict_document['section'],
            'question': dict_document['question']
        })
print(documents[20])
```

```python
index = Index(
    text_fields=["question","text","section"],
    keyword_fields=["course"]
)
index.fit(documents)
```

```python
query = "the course has already started, can I still enroll?"
query = "Can I join the course if it has already started?"
filter_dict = {"course": "data-engineering-zoomcamp"}
```

```python
results = index.search(
    query=query, 
    boost_dict={'question': 3.0, 'section':0.5},
    filter_dict=filter_dict,
    num_results=3
)
```
